## Merging the two datasets base on their Account ID

```python
import pandas as pd

def merge_datasets(dataset1_path, dataset2_path, merge_column):
    # Load the datasets
    df1 = pd.read_csv(dataset1_path)
    df2 = pd.read_csv(dataset2_path)
    
    # Merge the datasets based on the specified column
    merged_df = pd.merge(df1, df2, on=merge_column, how='inner')
    
    return merged_df

# Example usage:
dataset1_path = "../data/orders_data_competition.csv"
dataset2_path = "../data/clients_data_competition.csv"
merge_column = "Account ID"

df = merge_datasets(dataset1_path, dataset2_path, merge_column)
df.sample(10)
```

Since there is not too many null values, we can drop them 

```python
def drop_null_rows(df):
    # Drop rows with null values
    df.dropna(inplace=True)
    return df

# Example usage:
# Assuming df is your DataFrame
df = drop_null_rows(df)
```

```python
def convert_data_types(df):
    # Define data types dictionary
    data_types = {
        'Order ID': int,
        'Account ID': int,
        'Market Key': str,
        'Security ID': int,
        'Order Type': str,
        'Order Time': pd.Timestamp,
        'Order Via': str,
        'Is Completed': bool,
        'Is Canceled': bool,
        'Expire Date': pd.Timestamp,
        'Execution Status': str,
        'quantity': int,
        'Price': float,
        'Sector Name': str,
        'Executed Quantity': int,
        'Quantity': int,
        'Client ID': int,
        'Gender': str,  # Convert to boolean
        'Risk Rate': float,
        'Company Name': str,
        'Is Closed': bool,
        'Is Dormant': bool,
        'Is Profile Suspended': bool,
        'Is Client Suspended': bool,
        'Client Type Name': str,
        'OpenDate': pd.Timestamp,
        'BirthDate': pd.Timestamp
    }
    
    # Convert each column to its supposed data type
    for col, data_type in data_types.items():
        df[col] = df[col].astype(data_type)
    
    return df

# Example usage:
# Assuming df is your DataFrame
df = convert_data_types(df)
```

```python
# Data visualization
# Histogram of a numerical variable
plt.figure(figsize=(10, 6))
sns.histplot(df['numerical_column'], bins=20, kde=True)
plt.title('Histogram of Numerical Column')
plt.xlabel('Numerical Column')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of a numerical variable
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='numerical_column')
plt.title('Boxplot of Numerical Column')
plt.ylabel('Numerical Column')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
